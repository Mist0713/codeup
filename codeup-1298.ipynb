## 문제 개요
모범생 선우와 모범생 주원이는 여러분의 도움 덕에 ISEO에서 우수한 성적을 거둘 수 있었다! (3029번 문제 설명 참고)

한편 코드업 왕국에서 ISEO가 열릴 동안, 옆 나라 유클리드 왕국에서는 '정수해 빨리 찾기 대회'가 열리고 있었다.

이 대회의 우승자인 남우는 어떤 방정식이 주어져도 그 방정식의 정수해를 1초 내에 하나 이상 찾을 수 있는 엄청난 천재였다.

선우와 주원이가 ISEO에서 좋은 성적을 거두었다는 소식을 들은 남우는, 선우와 주원이에게 대결 신청을 하였다.

대결 내용은 1차 부정방정식 $Ax\,+\,By=\,C$의 정수해 $(x,y)$를 찾는 것으로, 더 빨리 찾은 쪽이 이기게 된다.

선우와 주원이는 남우보다 더 빨리 정수해를 찾아 남우와의 대결에서 이기고 싶다.

선우와 주원이를 도와줄 수 있도록, 방정식 $Ax\,+\,By=\,C$의 정수해를 1초 안에 찾아 주는 프로그램을 만들어 보자.

## 입력
첫 번째 줄에 $A, B, C$가 차례대로 공백으로 구분되어 주어진다.

$A,\,B,\,C$는 모두 $-10^9$이상 $10^9$ 이하의 정수이다.

## 출력
첫째 줄에 방정식 $Ax\,+\,By=\,C$를 만족하는 정수 $x,y$의 값을 <font color="red">아무거나</font> 공백으로 구분하여 출력한다.

만약 $Ax\,+\,By=\,C$를 만족하는 정수 $x,y$가 존재하지 않는다면 "Not Exist"를 출력한다.

## 풀이

선형 부정방정식이 정수해를 가지므로 먼저 이 방정식이 정수해를 가지는지 판별하기 위한 기준을 짜놓는다.

## 일차 선형 부정방정식이 정수해를 가질 조건
> $Ax+By=C (a\neq 0 또는 b \neq 0)$에서 $d = \gcd(a,b)$ 일 때, 선형 디오판토스 부정방정식이 정수해를 가질 조건은 $d|c$와 필요 충분 조건이다.

### 보조 정리 - 베주 항등식
>적어도 둘 중 하나는 0이 아닌 임의의 정수 $a,b$에 대해 $d=\gcd(a,b)$이면 정수 $u,v$가 반드시 존재하여 $$au+bv = d$$가 성립한다

### 증명
$(\Rightarrow)$ 임의의 정수해 $(x,y)$가 존재하면 $ax+by=c$의 좌변은 $d$로 나누어지므로 우변 $c$도 $d$로 나누어진다.

$(\Leftarrow)$ $au+bv=d$라는 임의의 해 $u,v$가 존재할 때, $d|c$ 이므로 $c = d\times k$인 임의의 정수 $k$가 존재한다. 이 때 $(x_0,y_0) = (ku,kv)

```python
def gcd(x, y):
    while y:
        x, y = y, x % y
    return abs(x)

def CheckIntegerAnswer(a,b,c): #0 : Infinite solutions, 1 : One solution, 2 : No solution
    if a == b == 0:
        if c == 0:
            return 0
        else:
            return 2
    elif c%gcd(a, b)==0:
        return 1
    else:
        return 2
```
